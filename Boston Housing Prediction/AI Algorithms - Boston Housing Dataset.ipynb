```python
from sklearn.datasets import load_boston
data = load_boston()
print (load_boston())
```

Print Headings and columns for data set

```python
data_frame = pd.DataFrame(data.data)
data_frame.columns = data.feature_names

data_frame.head(10)
```

Print a histogram of the quantity to predict: price



```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5, 5))
plt.hist(data.target)
plt.xlabel('price ($1000s)')
plt.ylabel('count')
plt.tight_layout()
```

Scatterplot For The Datatypes



```python
for index, feature_name in enumerate(data.feature_names):
    plt.figure(figsize=(4, 3))
    plt.scatter(data.data[:, index], data.target,c='indianred')
    plt.ylabel('Price', size=15)
    plt.xlabel(feature_name, size=15)
    plt.tight_layout()
```

Simple prediction model - LinearRegression



```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target)

from sklearn.linear_model import LinearRegression
clf = LinearRegression()
clf.fit(X_train, y_train)
predicted = clf.predict(X_test)
expected = y_test

plt.figure(figsize=(4, 3))
plt.scatter(expected, predicted, c = 'sandybrown')
plt.plot([0, 50], [0, 50])
plt.xlabel('Actual price ($1000s)')
plt.ylabel('Predicted price ($1000s)')
plt.tight_layout()
```
